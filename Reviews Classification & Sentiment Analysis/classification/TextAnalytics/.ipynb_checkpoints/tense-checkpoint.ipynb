*Tense classification of reviews using `nltk`*

```python
import pandas as pd
from nltk import *

rvws = pd.read_csv('Dataset3.csv')
```

```python
# import nltk
# nltk.download('averaged_perceptron_tagger')
```

```python
# tokenize review then tag each word using nltk's `pos_tag` function

def get_tags(x):
    tense = []
    only = []
    word = word_tokenize(x)
    tense.append(pos_tag(word))
    
    for t in tense:
        for i in t:
            only.append(i[1])
    return only

rvws['tag'] = rvws['processed_text'].map(lambda x: get_tags(x))
```

```python
# define functions to read each tag and identify how many words are future, past, or present tense

def future(x):
    future_tense = []
    for i in x:
        if i == "MD":
            future_tense.append(i)
            
    return len(future_tense)
            

def present_simp(x):
    present_tense = []
    for i in x:
        if i in ["VBP", "VBZ","VBG"]:
            present_tense.append(i)
        
    return len(present_tense)


def present_con(x):
    present_tense = []
    for i in x:
        if i.lower() in ["could", "would","should"] and ["if"]:
            present_tense.append(i)
        
    return len(present_tense)


def past(x):
    past_tense = []
    for i in x:
        if i in ["VBD", "VBN"]:
            past_tense.append(i)
        
    return len(past_tense)


    
rvws['future'] = rvws.tag.map(lambda x: future(x))
rvws['present_simple'] = rvws.tag.map(lambda x: present_simp(x))
rvws['present_con'] = rvws.processed_text.map(lambda x: present_con(x))
rvws['past'] = rvws.tag.map(lambda x: past(x))

rvws
```
