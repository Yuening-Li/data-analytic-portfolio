```python
# convert second set of labels to binary
rating = np.zeros(len(labels2))
usr_xp = np.zeros(len(labels2))
bug = np.zeros(len(labels2))
feature = np.zeros(len(labels2))


for x,i in zip(range(len(labels2)+1), labels2['label']):
    if i == 'rating':
        rating[x] = 1
    elif i == 'user experience':
        usr_xp[x] = 1
    elif i == 'feature request':
        feature[x] = 1
    elif i == 'bug report':
        bug[x] = 1
    else:
        print(labels2.iloc[x])
```
